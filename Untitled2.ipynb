Exercyce 8:

```python
x=int(input("Donner une valeur"))
y=int(input("Donner une autre valeur"))
z=x+y
if z==0:{
    print("la somme des deux valeur saisie est nul :" ,x, "+",y, "=", z)
}
elif z<0:{
    print("la somme des deux valeur saisie est Negatif :" ,x, "+",y, "=", z)
}
else:{
    print("la somme des deux valeur saisie est positif:",x, "+",y, "=", z)
}

```
