```python
!pip install turtle
import time
import datetime as dt
import tkinter as tk
t=tkinter.tkinter()
t1=tkinter.tkinter()
s=turtle.Screen()
s.bgcolor("white")
s = dt.datetime.now().second
mn = dt.datetime.now().minute
h = dt.datetime.now().hour
t1.pensize(3)
t1.color('black')
t1.penup()
t1.goto(-20, 0)
t1.pendown()
for i in range(2):
    t1.forward(200)
    t1.left(90)
    t1.forward(70)
    t1.left(90)
t1.tkinter()
t.clear()
while True:
    t1.tkinter()
    t.clear()
    t.write(str(h).zfill(2)
            +":"+str(mn).zfill(2)+":"
            +str(s).zfill(2),
            font =("Arial Narrow", 35,"bold"))
    time.slpeed(1)
    s+=1
    if s == 60:
        s == 0
        mn+=1
    if mn == 60:
        mn == 0
        h+=1
    if h == 23:
        h == 0
```
