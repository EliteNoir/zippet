```python
x=int(input("Entrer un nombre "))
mul=1
add=0
for i in range(1,x+1):
    ap=0
    for j in range(1,i+1):
        if i%j==0:
            ap=ap+ 1
    if ap==2:
        add=add+ i
    if i%2==0:
        mul=mul* i
print("L'addision des nombres premier est agale a ",add)
print("La multiplication des nombres paire est agale a ",mul)
```
