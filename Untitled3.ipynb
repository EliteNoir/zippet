```python
x=int(input("Entre un nombre "))
y=int(input("Entre un autre nombre "))
v=y*x
if v<0:{
    print("le resultat de l'opperation est negatif :",x, "*" ,y, "=" ,v)
}
elif v==0:{
     print("le resultat de l'opperation est nul :" ,x, "*" ,y, "=" ,v)
}
else :{
     print("le resultat de l'opperation est positif :" ,x, "*" ,y, "=" ,v)
}
```
