```python
text="ceci est un texte simple. le but de cet exercice est de calculer la frequence de chaque lettre dansn ce texte-ci"
def frequence(t):
    d={}
    for i in t:
        som=0
        for j in t:
            if i==j:
                som+=1
                d[i]=som
    print(d)           
frequence(text)       
```
