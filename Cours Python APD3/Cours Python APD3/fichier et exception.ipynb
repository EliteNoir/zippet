# Exception en python


Même si une instruction ou une expression est syntaxiquement correcte, elle peut générer une erreur lors de son exécution. Les erreurs détectées durant l'exécution sont appelées des exceptions et ne sont pas toujours fatales : nous apprendrons bientôt comment les traiter dans vos programmes.

```python
a = 10
b = 2

print(a/b)
```

```python
a = 10
b = 0

print(a/b)
print("End Code")
```

```python
#Try…Except

a = 10
b = 0

try:
    print(a/b)
except Exception:
    print("donnée diviseur différent de zero")
```

```python
a = 10
b = 0

try:
    print(a/b)
except ZeroDivisionError:
    print("You cannot divide a number by zero")
except Exception:
    print("Unknown Error occurred")

print("End Code")
```

```python
#Try…Except…Finally
try:
    f = open("example.txt", 'w')
    f.write("Test")
    print("Write operation was successful")
except Exception:
    print("Something went wrong...")
finally:
    f.close()
```

```python
#Try…Except with Else
num = 12

try:
    if num != 0:
        print("Non zero")
except Exception:
      print("Something went wrong")
else:
      print("No errors found")
```
