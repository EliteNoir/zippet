```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt

text = "keys  Karim  development Python webscraping automation getsetpython projects cool Maureen Okolo awesome blockchain data machinelearning  regex Sawadogo Ibro datascience opencv ML list dictionary analysis sentiment variable set abiodounkpevi  enumerate try except if-else for function  object instance GUI encryption security"

cloud = WordCloud(background_color="red").generate(text)

plt.imshow(cloud)
plt.axis('off')
plt.show()
```
