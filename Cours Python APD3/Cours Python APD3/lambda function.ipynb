Les lambdas Python sont de petites fonctions anonymes, soumises Ã  une syntaxe plus restrictive mais plus concise que les fonctions Python normales.

```python
# Definie la funtion lambda
lambda x: x**2
```

```python
squared = lambda x: x**2
squared(5)
```

```python
puissance = lambda x, y: x**y
puissance(2,5)
```

```python
def squared_1(x):
    return x**2
squared_1(5)
```

```python
# Mettre des argument dans la fonction lambda
def func(num,calback_fun):
    return calback_fun(num)


func(2,squared_1)
```

```python
(lambda x: x**2)(3)
```

```python
def num_pairs(num):
    if num % 2 == 0:
        print('pair')
    else:
        print('Impair')
num_pairs(2)
```

```python
def num_pairs1(num):
    return 'pair' if num % 2 == 0 else 'impair' 
```

```python
num_pairs1(3)
```

```python
pair_impair = lambda num: 'pair' if num % 2 == 0 else 'impair'
```

```python
pair_impair(5)
```
