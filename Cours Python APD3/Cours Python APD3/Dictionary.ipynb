# Dictionnaire en python

Un dictionnaire en python est une sorte de liste mais au lieu d'utiliser des index , on utilise des clés alphanumériques.

Comme les listes, les dictionnaires permettent de "stocker" des données. Chaque élément d'un dictionnaire est composé de 2 parties, on parle de paires **"clé/valeur"**. 

On précise entre guillemets la clé, le signe deux points « : » et la valeur correspondante. On sépare les différents couples clé : valeur par une virgule. C'est d'ailleurs comme cela que Python vous affiche un dictionnaire quand vous le lui demandez.

```python
couleur = {'bleu':10,'orange':6,'gris':9}
print(couleur)
```

```python
print(type(couleur))
```

```python
# acceder  a la valeur du dict

couleur['bleu']
couleur['orange']
```

```python
couleur1 = dict([('bleu',10),('orange',6),('gris',10)])
print(couleur1)
```

```python
couleur['bleu'] = 11
print(couleur)
```

# Les methodes sur le dictionaire

```python
couleur.items()
```

```python
list(couleur.items())
```

```python
#Pour récupérer les clés on utilise la méthode keys .
couleur.keys()
```

```python
#Pour récupérer les valeurs on utilise la méthode keys .
couleur.values()
```

```python
print(list(couleur.keys()),list(couleur.values()))
```

```python
couleur.update({'jaune':7}) #Ajouter un element dans le dict
print(couleur)
```

```python
couleur.popitem()
print(couleur)
```

```python
couleur.pop('bleu')
```

```python
print(couleur)
```

```python
del couleur['gris']
print(couleur)
```
